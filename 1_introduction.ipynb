### Read Data Files in Python

Of course, one of the most pressing issues in learning Python is how we are supposed to read in data before analysis?

We will cover this issue step by step.

Let's assume that the data is stored in an Excel spreadsheet, the most common scenario.

1.   Suppose the Excel file is directly available on the internet, like many of those on the Stock and Watson [textbook companion website](https://www.pearsonhighered.com/stock_watson).

  For example, here's the direct [link](https://media.pearsoncmg.com/ph/bp/bp_stock_econometrics_4_cw/content/datapages/data/Age_HourlyEarnings.xlsx) for a data set intended for the Empirical Exercise 2.1 in the textbook. If you want to directly load the data contained in the Excel file into the current Python session for analysis, do this:

```python
import pandas

# Reading an Excel file
dataset = pandas.read_excel('https://media.pearsoncmg.com/ph/bp/bp_stock_econometrics_4_cw/content/datapages/data/Age_HourlyEarnings.xlsx')
dataset.head()
```

In the above code, `pandas` is a library that packages fast, efficient data analysis tools for Python. Along with `numpy`, We will use `pandas` a lot for this course.

A careful look reveals that the first two rows and the first column are actually not part of the data. They are labels of the rows and the columns. If we want to exclude them from showing up in the resulting import, we could do this instead (since you already loaded the `pandas` library, you do not have to do it again in the same session; but you do need to import it again in a session anew):

```python
# Reading an Excel file
dataset = pandas.read_excel('https://media.pearsoncmg.com/ph/bp/bp_stock_econometrics_4_cw/content/datapages/data/Age_HourlyEarnings.xlsx', skiprows=2, header=None, usecols=list(range(1, 11)), nrows=29)
dataset.head()
```

You can always use `help()` command to look at the syntax explanation of a particular command or function (not that it is easily understandable though):

```python
help(pandas.read_excel)
```

(We will talk about the data structure in general and in particular after reading an Excel file by using `pandas` commands. For now, the focus is on how we can read in the data from the Excel file.)

2.   What if your Excel data file is on your local computer? For example, you may have downloaded it from some where previously?

There are more than one way of doing it. That said, the easiest way I found, is to upload the file into the current Python session. The updated file will not be located in the same folder as is the current Jupyter Notebook file on your Google Drive. In fact, it will be located in a temporary folder temporarily, and after the current Python session closes, the file will be deleted from the cloud.

You can achieve it by this:

```python
from google.colab import files
uploaded = files.upload()
```

After uploading your file following the prompt, you can simply read in the Excel file by directly quoting its name.

```python
import pandas
dataset = pandas.read_excel('Age_HourlyEarnings.xlsx', skiprows=2, header=None, usecols=list(range(1, 11)), nrows=29)
dataset.head()
```

Now, even the above code is more complicated than necessary, because it involves two steps: (1) you have to locate the file on your computer to upload; (2) you have to specify the file name (in this case `Age_HourlyEarnings.xlsx`) for Python to read in. Assuming that the computer and Python are super smart, should not we expect them to automatically remember the file name when uploading, therefore to save our effort of typing in the file name mannually?

Exactly. So this is what the following code is doing (the `file_path` line of code remembers where the file is uploaded and what the file name is and is used in the next line):

```python
from google.colab import files
import pandas

uploaded = files.upload()
file_path = list(uploaded.keys())[0]
dataset = pandas.read_excel(file_path, skiprows=2, header=None, usecols=list(range(1, 11)), nrows=29)

dataset.head()
```

So, where is the uploaded file exactly located? It will be saved in the same temporary folder mentioned above, but, honestly, I don't know and I don't care, provided that Google, as it promises, will delete the file after the Colaborary session is done.
