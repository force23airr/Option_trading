```python
import numpy
import pandas
import scipy.stats
# import statsmodels modules that enable us to implement OLS
import statsmodels.formula.api as smf
import statsmodels.api as sm
from google.colab import files
```

Read in the data [hmda_sw.xlsx](https://docs.google.com/spreadsheets/d/1FoQSCmW17EobJltIZmTH5ha5MavWK3PM/edit?usp=drive_link&ouid=110351823593825164156&rtpof=true&sd=true).

```python
uploaded = files.upload()
file_path = list(uploaded.keys())[0]
dataset = pandas.read_excel(file_path)
# if there is any error in this step (pandas.read_excel)
# try opening the Excel file on your local computer and saving it again.

dataset.head()
```

```python
dataset['deny']=(dataset['s7'] == 3).astype(int)
dataset['pi_rat'] = dataset['s46'] / 100
dataset['black'] = (dataset['s13'] == 3).astype(int)
```

```python
mod = smf.ols(formula='deny ~ pi_rat', data=dataset)
result = mod.fit()
result2 = result.get_robustcov_results(cov_type='HC1')
print(result2.summary())
```

```python
mod = smf.ols(formula='deny ~ pi_rat+black', data=dataset)
result = mod.fit()
result2 = result.get_robustcov_results(cov_type='HC1')
print(result2.summary())
```

```python
mod = smf.probit('deny ~ pi_rat', data=dataset)
result = mod.fit()
print(result.summary())
```

```python
mod = smf.probit(formula='deny ~ pi_rat', data=dataset)
result = mod.fit(cov_type='HC1')
print(result.summary())
```

```python
hypotheses = 'pi_rat = 3.0'
test_result = result.t_test(hypotheses)
print(test_result)
```

```python
predicted_probability = result.predict(pandas.DataFrame({'pi_rat':[0.3]}))
print(predicted_probability)
```

```python
predicted_probability = result.predict(pandas.DataFrame({'pi_rat':[0.4]}))
print(predicted_probability)
```

```python
mod = smf.logit(formula='deny ~ pi_rat', data=dataset)
result = mod.fit(cov_type='HC1')
print(result.summary())
```

```python
mod = smf.probit('deny ~ pi_rat+black', data=dataset)
result = mod.fit(cov_type='HC1')
print(result.summary())
```

```python
hypotheses = 'pi_rat = 3.0, black = 0.0'
test_result = result.f_test(hypotheses)
print(test_result)
```

```python
predicted_probability = result.predict(pandas.DataFrame({'pi_rat':[0.3], 'black':[0]}))
print(predicted_probability)
```
