### Pandas ###

Pandas is a package of fast, efficient data management tools for Python. It is very flexible too. Again, before using it, you have to load it first.

(In fact, whenever we import `pandas`, we often import `numpy` too, because `pandas` is built upon `numpy` infrastructure and you never know whether and when you will need `numpy`.)

```
import numpy
import pandas
```

**DataFrame** is of several columns, one for each variable.
(There are other types of data `pandas` deals with; we'll cover them when we see them.)

Now, use `read_csv` to read a data set of spreadsheet in csv format (you can use other corresponding facilities to read excel, stata, text formats; for example, to read stata file, use `read_stata`; and we already saw how to read Excel data file.)

```
dataset = pandas.read_csv('https://github.com/QuantEcon/QuantEcon.lectures.code/raw/master/pandas/data/test_pwt.csv')
```

```
dataset.head()
```

You see those bold words and numbers on the top row and the left-most column: they are column labels and index numbers, respectively. In this case, the top row labels are already in the original `csv` data file, but the indices are auto-generated by Python.

```
# Show particular rows
dataset[2:5]
# Remember the general rule: a[m:n] returns n - m elements, starting at a[m].
# The same rule goes with dataframe rows.
```

```
# show particular columns
dataset[['country','year','cg']]
```

```
# show particular rows and columns
dataset.iloc[2:5,[0,2,3]]
```

```
dataset.loc[dataset.index[2:5], ['country','year','POP']]
```

```
# suppose we only want to analyze the data of above variables
dataset2 = dataset[['country','year','POP']]
```

```
dataset2
```

Now, back to the data frame `dataset`, could we filter the data according to some criteria, only preserving the rows that have `POP > 10000` and `cc > 70`, for example?

The answer is yes. There're multiple ways of doing this, but one of them is to use `numpy` `where` function to retrieve the indices of the rows that satisfy those criteria, and then to retrieve the data for these rows.

```
idx = numpy.where( (dataset['POP']>10000) & (dataset['cc']>70) )
dataset.loc[idx]
```

```
dataset3=dataset[(dataset['POP']>10000) & (dataset['cc']>70)]
```
